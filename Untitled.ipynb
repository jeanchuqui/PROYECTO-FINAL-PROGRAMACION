```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
output_notebook()
archivo = pd.read_csv("datos.csv",encoding="ISO-8859-1")
source = ColumnDataSource(archivo)
archivo
```

```python
edad, age = np.unique(archivo['Edad'], return_counts=1)
plano1 = figure(plot_width=1000, plot_height=500, title='Asesinatos por edad');
plano1.title.text_font_size = "20pt"
plano1.vbar(x=edad, width=0.3, top=age)
show(plano1)
```

```python
hora_muerte,f_hmuerte = np.unique(archivo['Hora Infracci¢n'].astype(str), return_counts=1)
plano2 = figure(y_range=hora_muerte,plot_width=800,plot_height=3000,title='Cantidad de muertes por hora')
plano2.title.text_font_size = '20pt'
plano2.hbar(y=hora_muerte,height=0.3,right=f_hmuerte) 
show(plano2)
```

```python

h_muerte,f_muerte = archivo['Hora Infracci¢n'].astype(str), archivo['Fecha Infracci¢n'].astype(str)
h_muerte_range, fh_muerte_range = np.unique(h_muerte, return_counts=1)
f_muerte_range, ff_muerte_range = np.unique(f_muerte, return_counts=1)
plano3 = figure(x_range=h_muerte_range,y_range=f_muerte_range,plot_width=4500,plot_height=5000,title='Fecha de muertes por hora')
plano3.title.text_font_size = '20pt'
plano3.yaxis.major_label_text_font_size = '10pt'
plano3.xaxis.major_label_text_font_size = '10pt'
plano3.yaxis.major_label_orientation = -0.2
plano3.xaxis.major_label_orientation = 0.8
plano3.circle(x=h_muerte,y=f_muerte, radius=0.356)
show(plano3)
```

```python
etnia,f_etnia = np.unique(archivo['Nacionalidad'],return_counts=1)
plano4 = figure(x_range=etnia,plot_width=2000,plot_height=800,title='Muertes de inmigrantes')
plano4.title.text_font_size = '20pt'
plano4.xaxis.major_label_orientation=0.4
plano4.vbar(x=etnia, width=0.3,top=f_etnia)
show(plano4)
```
